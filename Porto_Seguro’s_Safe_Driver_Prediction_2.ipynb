```
var_list = train.columns.tolist()[2:]
```

```
for i in var_list :
  print(len(train[i].unique()))
```

```
des_list = []
real_list = []
for i in var_list :
  if len(train[i].unique()) < 105 :
    des_list.append(i)
  else :
    real_list.append(i)
```

```
print(des_list)
print(real_list)
```

```
for i in des_list :
  print(pd.crosstab(index = train['target'], columns = train[i]))
```

```
# chi-square independence test

des_bad_list = []
des_good_list = []

import scipy as sp
for i in des_list :
  obs = np.array(pd.crosstab(index = train[i], columns = train['target']))
  pvalue = sp.stats.chi2_contingency(obs)[1]
  if pvalue > 0.05 :
    print("%s seems useless with p-value : %s" %(i, pvalue))
    des_bad_list.append(i)
  else :
    print("*****%s seems useless with p-value : %s*****" %(i, pvalue))
    des_good_list.append(i)
```

```
print(len(des_good_list), len(des_bad_list))
```

```
# ANOVA with real-valued variables

import statsmodels.api as sm
from statsmodels.formula.api import ols
  
for i in real_list :
  lm1=ols("%s ~ C(target)" %(i), data = train).fit()
  table = sm.stats.anova_lm(lm1, typ=2)
  print(table)
```

```
# we'll do one-hot with des_good_list
full = pd.concat([train, test], axis = 0)
full_2 = pd.get_dummies(full, prefix=des_good_list, columns = des_good_list)
for i in des_bad_list :
  full_2.drop(columns = [i], inplace = True)  
train = full_2.iloc[0 :len(train)]
test = full_2.iloc[len(train) : ]
del full, full_2
```

# Logistic Regression

```
train = train.dropna()
```

```
# log-reg first

from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

test_log = test['target']
model = LinearRegression()
model.fit(train[var_list], train['target'])
test_log = model.predict(test[var_list])
print(r2_score(train['target'], model.predict(train[var_list]).round()))
```

```
test_log = pd.concat([test['id'], pd.DataFrame(test_log)], axis = 1)
test_log = test_log.rename(columns = {0: "target"}) 
test_log.to_csv('logistic_pred2.csv', index = False)
```

0.23................

# Random Forest

```
from sklearn.ensemble import RandomForestRegressor

test_ran = test[['id', 'target']]
model = RandomForestRegressor(n_estimators=200, max_depth=10)
model.fit(train[var_list], train['target'])
test_ran['target'] = model.predict(test[var_list])
print(r2_score(train['target'], model.predict(train[var_list]).round()))
```

# LGBM

```
import lightgbm as lgb
params = {}
params['learning_rate'] = 0.01
params['boosting_type'] = 'gbdt'
params['objective'] = 'binary'
params['metric'] = 'binary_logloss'
params['sub_feature'] = 0.9
params['num_leaves'] = 50
params['min_data'] = 50
params['max_depth'] = 10

test_lgbm = test[['id', 'target']]
d_train = lgb.Dataset(train[var_list], label=train['target'])
model = lgb.train(params, d_train, 200)
test_lgbm['target'] = model.predict(test[var_list])
print(r2_score(train['target'], model.predict(train[var_list]).round()))
```
